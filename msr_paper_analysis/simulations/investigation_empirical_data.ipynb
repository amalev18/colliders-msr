```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Reading the file containing all features used in analysis of the found paper
all_features = pd.read_csv('../empirical_analysis/data/commit_all_features.csv')

# Extract features to examine
columns = ['age', 'leftLocationCount', 'leftLocationDiff', 'rightLocationCount', 'rightLocationDiff', 'dev_ref_com_exp', 
           'dev_ref_exp', 'num_touched_before', 'refactoryCount', 'refactoryTypeCount','codeElementCount', 'fileTouchCount',
           'fileTouchAverage']
```

```python
# Looking at statistics of each feature, and plotting histograms, in order to retrieve information about the characteristrics 
# present in the real data.

for column in columns:    
    feature = all_features[column]

    feature_sorted = feature.sort_values()

    percentile_75 = feature_sorted.quantile(0.75)
    median = feature_sorted.median()
    mean = feature_sorted.mean()
    min = feature.min()
    max = feature.max()
    output_str = f'Feature: {feature.name}, Third Quartile: {percentile_75}, Median: {median}, Mean: {mean}, Min: {min}, Max: {max}'
    print(output_str)
    
    plt.hist(feature, bins=80, edgecolor='black')

    plt.xlim(0, max)
    plt.xlabel(f'{feature.name}')
    plt.ylabel('Density')
    plt.title('Density Plot')

    plt.show()
```
